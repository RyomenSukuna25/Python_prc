```python
import ipywidgets as widgets
from IPython.display import display

def submit_callback(change):
    name = name_entry.value.strip()
    if not name:
        print("Error: Please enter your name.")
        return
    gender = gender_dropdown.value
    skills = []
    if skill_python.value:
        skills.append("Python")
    if skill_java.value:
        skills.append("Java")
    if skill_cpp.value:
        skills.append("C++")
    skill_text = ", ".join(skills) if skills else "None"
    print(f"\nInformation Submitted:\nName: {name}\nGender: {gender}\nSkills: {skill_text}\n")

def custom_dialog_callback(change):
    print("Warning: This is a custom warning dialog!")

name_entry = widgets.Text(placeholder='Enter your name')
gender_dropdown = widgets.Dropdown(options=['None', 'Male', 'Female'], value='None', description='Gender:')
skill_python = widgets.Checkbox(value=False, description='Python')
skill_java = widgets.Checkbox(value=False, description='Java')
skill_cpp = widgets.Checkbox(value=False, description='C++')
submit_button = widgets.Button(description='Submit')
dialog_button = widgets.Button(description='Show Custom Dialog')
submit_button.on_click(submit_callback)
dialog_button.on_click(custom_dialog_callback)

display(widgets.VBox([
    widgets.Label("User Information Form", style={'font_weight': 'bold'}),
    name_entry,
    gender_dropdown,
    widgets.Label("Select Skills:"),
    skill_python,
    skill_java,
    skill_cpp,
    submit_button,
    dialog_button
]))
```
